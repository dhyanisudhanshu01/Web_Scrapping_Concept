This is the Web Scraping of Meesho website. In this we are going to extract the List of shoes and show and store in Dataframe and save it in csv.

Import The required libraries

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

Load the webpage, find the elements and save the html in a file using Selenium 

```python
driver = webdriver.Chrome()
query = "shoes"

driver.get(f"https://www.meesho.com/search?q={query}&searchType=manual&searchIdentifier=text_search")
elems = driver.find_elements(
By.CSS_SELECTOR,
    ".sc-dQkurY.ProductListItem__GridCol-sc-1baba2g-0.bwMLBY.eucHxL"
    )
print(f"{len(elems)}items found")
with open(f"../data/{query}.html", "w", encoding="utf-8") as f:
    for elem in elems:
        a = elem.get_attribute("outerHTML")    
        f.write(a+"\n")
time.sleep(2)
driver.close()
```

Collecting the data using Beautiful Soup

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Read the saved HTML file and parse it with BeautifulSoup
with open("../data/shoes.html", "r", encoding="utf-8") as f:
    html = f.read()
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
items = soup.find_all("div", class_="sc-dQkurY ProductListItem__GridCol-sc-1baba2g-0 bwMLBY eucHxL")
```

```python
# Extract the title and price of each item and store them in a dictionary
d = {"title": [], "price": []}
for item in items:
    title = item.find("p", class_="sc-dOfePm ivsDoe NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ffaqxo").text.strip()
    price = item.find("h5", class_="sc-dOfePm jSZBdj").text.strip()
    d["title"].append(title)
    d["price"].append(price)
```

```python
# Create a DataFrame from the dictionary and display it
df = pd.DataFrame(d)
print(df)
```

```python
# Save the DataFrame to a CSV file
df.to_csv("../data/shoes.csv", index=False)
```
