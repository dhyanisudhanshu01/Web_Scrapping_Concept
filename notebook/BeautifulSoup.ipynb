This is the Web Scraping of Tata IPL Auction 2026 website. In this we are going to extract the Auction overview grid view and show and store in Dataframe and save it in csv.

```python
#Importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Calling the URL and getting the response
url = "https://www.iplt20.com/auction/2026"
response = requests.get(url)
print(response)
```

```python
#Parsing the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
print(soup)
```

```python
#Finding the div with class 'auction-grid-view mt-3 overview-2026' which has the overview of the auction
auction_overview = soup.find('div', class_='auction-grid-view mt-3 overview-2026')
print(auction_overview)
```

```python
#Finding all the divs with class 'agv-main' which has the team names
team_list = auction_overview.find_all('div', class_='agv-main')
print(team_list)
```

```python
#Extracting the Columns from div
teams_overview = []
for team in team_list:
    team_name = team.findAll('span', class_='fr-name')
    for header in team_name:
        name = header.text.strip()
        teams_overview.append(name)
```

```python
teams_overview = list(set(teams_overview))
print(teams_overview)
```

```python
print(teams_overview)
```

```python
#Extracting the values from the div and creating a dataframe
values = []
for team in team_list:
    team_dict ={}
    team_dict["Team"] = team.get('id')
    team_value = team.findAll('span', class_='fr-fund')
    team_name = team.findAll('span', class_='fr-name')
    for name, value in zip(team_name, team_value):
        name = name.text.strip()
        team_dict[name] = value.text.strip()
    values.append(team_dict)
df = pd.DataFrame(values)
    
```

```python
#Saving the dataframe to a csv file
df.to_csv('IPL_Auction_2026.csv', index=False)
```
